```python
from requests_html import HTMLSession
```

```python
session = HTMLSession()
```

```python
url="http://ihadis.com/"
r = session.get(url)
pattern = "body > div.main-wrap.p-bottom > div.header_area > div > div.col-xs-12.book-list-small > div.hadith-list"
book_details = r.html.find(pattern)[0].text
# print("details: ", len(book_details),type(book_details))
```

```python
book_details_list=book_details.split("\n")
book_number = book_details_list[0::3]
book_names = book_details_list[1::3]
# print(book_names[0])
# print(len(book_names))
hadith_number = book_details_list[2::3]
# print(len(book_number),len(book_names),len(hadith_number))
```

```python
def hadith_book_scrapping(book_url,book_name):
  r = session.get(book_url)
  chapter_pattern = "body > div.main-wrap.p-bottom > div.container.main-body > div > div.col-md-9.col-xs-12.rl > div.hadith-list"
  details = r.html.find(chapter_pattern)[0].text
  details_list=details.split("\n")
  total_chapter = details_list[0::3]

  for i in range(1, len(total_chapter)+1):
    # print(book_url)
    chapter_url = book_url + "/chapter/" + str(i)
    print("chapter_url:",chapter_url)

    r = session.get(chapter_url)
    # define the pattern
    no_pattern = "h3.hidden-print"
    arabic_text_pattern = "p.hadith-des2"
    bangla_text_pattern = ".hadith-des"
    narrator_pattern = ".narrated-by"
    hadith_status_pattern = ".label.validity"

    numbers = [i.text for i in r.html.find(no_pattern)]
    arabic_text = [i.text for i in r.html.find(arabic_text_pattern)]
    bangla_text = [i.text for i in r.html.find(bangla_text_pattern)]
    narrator = [i.text for i in r.html.find(narrator_pattern)]
    hadith_status = [i.text for i in r.html.find(hadith_status_pattern)]
    print("book:",book_name,"Chapter: ", i, len(numbers), len(arabic_text), len(bangla_text), len(narrator), len(hadith_status))

    with open(f"{book_name}_chapter_{i}.txt", "w") as f:
      f.write("No., Arabic Text, Bangla Text, Narrator, Hadith Status\n")
      for n, a, b, na, hs in zip(numbers, arabic_text, bangla_text, narrator, hadith_status):
        f.write(f"{n}, {a}, {b}, {na}, {hs}")
        f.write('\n')
```

```python
import time
book_details_link2 = r.html.find("a")
cnt=0
for i in range(7,37):
  start = time.time()
  book_url=book_details_link2[i].attrs["href"]
  book_name=book_names[cnt]
  # print("i=",i,book_url,book_name)
  cnt+=1
  hadith_book_scrapping(book_url,book_name)

print(f"=========={time.time()-start}=======")
```
